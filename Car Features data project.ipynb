Our goal is to explore the relationship between these features and the price of the car, and to gain insights that can help us better understand the factors that influence car prices.

Methods:

Data exploration: We will use descriptive statistics and data visualizations to explore the distribution of the features and the relationships between the features and price.

Regression analysis: We will use regression analysis to model the relationship between the features and price, and to make predictions about car prices based on the features.
Expected outcomes:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Car Features and MSRP.csv')
data
```

To display first few rows of our data

```python
data.head()
```

To display the bottom rows

```python

data.tail()
```

Exploring our dataset by getting a summary of the data

```python
print(data.info())
print(data.describe())
```

Dropping irrelevant columns. These are columns that have no impact in my analysis:

```python
data=data.drop(['Engine Fuel Type','Market Category','Vehicle Style','Popularity','Number of Doors','Vehicle Size'],axis=1)
data.head()
```

Dropping duplicate rows... Data shape before:

```python
#Total no of rows and columns
data.shape

```

Rows with duplicate data:

```python
duplicate_rows_data=data[data.duplicated()]
print("no of duplicate rows:", duplicate_rows_data.shape)
```

```python
data.count()
```

```python
#Dropping duplicates
data=data.drop_duplicates()
data.head()
```

Counting the number of rows after removing duplicates

```python
data.count()
```

Finding null values... The null values present in Engine HP and Engine Cylinders explain why in the data count above the two have different values than the rest

```python
print(data.isnull().sum())
```

```python
#Dropping the null values
data=data.dropna()
data.count()
```

```python
#Double checking to make sure the null values have been removed.
#Null value count
print(data.isnull().sum())
```

HISTOGRAM

```python
plt.hist(data['city mpg'], bins=20)
plt.xlabel('city mpg')
plt.ylabel('Count')
plt.title('Distribution of City MPG')
plt.show()
```

SCATTER PLOT

```python
plt.scatter(data['Engine HP'], data['MSRP'])
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.title('Engine HP vs. MSRP')
plt.show()
```

BOX PLOT

```python
sns.boxplot(x='Engine Fuel Type', y='highway MPG', data=data)
plt.xlabel('Engine Fuel Type')
plt.ylabel('highway MPG')
plt.title('highway MPG by Engine Fuel Type')
plt.show()
```

PAIR PLOT

```python
sns.pairplot(data[['Engine HP', 'highway MPG', 'MSRP']])
plt.show()
```
